```python
from collections import Counter

def read_lines_in_chunks(file, lines_per_chunk=10000):
    chunk = []
    for line in file:
        chunk.append(line.strip())
        if len(chunk) >= lines_per_chunk:
            yield chunk
            chunk = []
    if chunk:
        yield chunk

chunk_counts = []

def process_data(chunk):
    error_counts = dict(Counter(chunk))
    chunk_counts.append(error_counts)

with open("logs.txt", "r", encoding="utf-8") as f:
    for chunk in read_lines_in_chunks(f):
        process_data(chunk)
```

### exe2

```python

for i, counts in enumerate(chunk_counts):  
    print(f'Chunk {i+1}:') 
    for key, value in counts.items():
        error = key.split("Error: ")[-1].split('"')[0]  
        print(f'  Error: {error}, Count: {value}')
    
```

### exe3


```python


total_counts = Counter()


for error_counts in chunk_counts:
    total_counts.update(error_counts)

print("Total Counts:")
for key, value in total_counts.items():
    error = key.split("Error: ")[-1].split('"')[0]  
    print(f'  Error: {error}, Count: {value}')
```

### exe4

```python
def findFreq(n):
    top_n = total_counts.most_common(n)
    print(f"{n} common errors:")
    for error, count in top_n:
        print(f" {error.split("Error:")[-1].split('"')[0] }: {count}")
findFreq(3)
```

```python

from pathlib import Path
file_path_xlsx = Path(r'time_series.xlsx')
file_path_parquet = Path(r'time_series.parquet')

def open_file(file):
    if file.suffix.lower() == '.xlsx':
        return  pd.read_excel(file)
    elif file.suffix.lower() == '.parquet':
        return pd.read_parquet(file)
        
df = open_file(file_path_parquet)
df
```

```python
df
```

```python

dup = df[df.duplicated()]
if not dup.empty:
    print(dup)
    df=df.drop_duplicates()

```

```python
df.loc[:, 'timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

invalid_dates = df[df['timestamp'].isna()]
if not invalid_dates.empty:
    print("there is invalid dates")
    print(invalid_dates)
df = df.dropna(subset=['timestamp'])
```

```python
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df['value']=df['value'].fillna(df['value'].mean()) 

df
```

### exe2.1

```python
df = df.copy()
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df['hour']=df['timestamp'].dt.floor('h')
df
```

```python
df.groupby('hour')['value'].mean().reset_index()
```

### exe2.2

```python
df['day'] = df['timestamp'].dt.day  
groups = {day: group for day, group in df.groupby('day')}  


print(groups[1].head())
```

```python
results = []
for g in groups.values():
    results.append(g.groupby('hour')['value'].mean().reset_index())
    
for r in results:
    print(r)
```

```python
final_df = pd.concat(results, ignore_index=True)  
print(final_df)
```
